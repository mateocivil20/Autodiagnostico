```python
columnas=data.columns.tolist()
```

```python
col_noPreg=['respondent_id','collector_id','date_created','date_modified','ip_address','email_address',
 'first_name','last_name','custom_2','PAIS','PAIS2','CIUDAD','CIUDAD 2','GRUPO','INICIO O FIN',
 'NOMBRE EMPRESA','SECTOR ECONÓMICO','NOMBRE','EMAIL','TIPO DE IDENTIFICACIÓN','NUMERO IDENTIFICACIÓN',
 'GENERO','FECHA NACIMIENTO','NIVEL JERARQUIA','NIVEL JERARQUIA2','CARGO','ANTIGÜEDAD','NIVEL EDUCACION',
 'SEGUNDO INDIOMA','SEGUNDO IDIOMA2','DOMINIO','AREAS MAYOR INNOVACION','AREAS MAYOR INNOVACION2']
```

```python
data_melted = pd.melt(data, id_vars=col_noPreg, var_name='Pregunta', value_name='Valor')
```

```python
data_melted = data_melted.merge(preg, on='Pregunta', how='left')
```

```python
data_melted.head()
```

```python
#Corregimos la jerarquia
jerarquia=[]

for i in range(0,len(data_melted)):
    if (data_melted['NIVEL JERARQUIA'][i] !='') or (data_melted['NIVEL JERARQUIA'][i] !=' ') :
        jerarquia.append(data_melted['NIVEL JERARQUIA'][i])
    else:
        jerarquia.append(data_melted['NIVEL JERARQUIA2'][i])

data_melted['jerarquia_final'] = jerarquia
```

```python
ruta_salida = r'C:\Users\mateo\Documents\Uniandes\github\Autodiagnostico\data\data.xlsx'
data_melted.to_excel(ruta_salida, index=False) 
```
